```python
import ROOT
from pyamptools import atiSetup
import os
os.chdir("/work/halld/home/dbarton/gluex/KShortPipLambda/sdme")
os.listdir()
atiSetup.setup(globals(), use_fsroot=True)

fileName = "/work/halld/home/dbarton/gluex/KShortPipLambda/sdme/twopi_plot.root"

def list_trees_and_branches(filename):
    file = ROOT.TFile.Open(filename)
    keys = file.GetListOfKeys()

    for key in keys:
        name = key.GetName()
        obj = key.ReadObj()
        print(f"{name:30} --> {obj.ClassName()}")

    file.Close()

list_trees_and_branches(fileName)
```
